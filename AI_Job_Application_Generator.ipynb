# AI-Powered Personalized Job Application Generator using RAG

## Objective

This project uses Retrieval-Augmented Generation (RAG) to generate personalized job application emails and cover letters.

Given:
- A candidate's resume
- A job description

The application:
1. Extracts skills and projects from the resume.
2. Stores them as embeddings in ChromaDB.
3. Retrieves the most relevant information based on the job description.
4. Uses an LLM to generate a personalized application email.

```python
from pypdf import PdfReader

reader = PdfReader("ashu (5).pdf")

print("Total Pages:", len(reader.pages))
```

```python
resume_text = ""

for page in reader.pages:
    resume_text += page.extract_text()

print(resume_text)
```

```python
print("Characters:", len(resume_text))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

print("✅ Text Splitter imported successfully")
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)
```

```python
chunks = text_splitter.split_text(resume_text)
```

```python
print("Total Chunks:", len(chunks))
```

```python
for i, chunk in enumerate(chunks):
    print("=" * 50)
    print(f"Chunk {i+1}")
    print("=" * 50)
    print(chunk)
    print()
```

# Step 5: Generate Embeddings

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embeddings = model.encode(chunks)
```

```python
print("Number of Chunks:", len(chunks))
print("Embedding Shape:", embeddings.shape)
```

```python
print(embeddings[0])
```

```python
print(len(embeddings[0]))
```

```python
embeddings.shape
```

# Step 6: Create Vector Database

```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.create_collection(
    name="resume_collection"
)
```

```python
for i, chunk in enumerate(chunks):
    collection.add(
        ids=[str(i)],
        documents=[chunk],
        embeddings=[embeddings[i].tolist()]
    )
```

```python
print("Total Documents:", collection.count())
```

```python
data = collection.get()

print(data.keys())
```

```python
for doc in data["documents"]:
    print("="*50)
    print(doc)
```

```python
collection.count()
```

# Step 7: Semantic Search

```python
job_description = """
We are looking for a Machine Learning Intern.

Requirements:

- Python
- TensorFlow
- Deep Learning
- CNN
- Computer Vision
- Problem Solving
"""
```

```python
job_embedding = model.encode(job_description).tolist()
```

```python
results = collection.query(
    query_embeddings=[job_embedding],
    n_results=3
)
```

```python
print(results)
```

```python
print("Most Relevant Resume Sections:\n")

for i, doc in enumerate(results["documents"][0], start=1):
    print("=" * 60)
    print(f"Result {i}")
    print("=" * 60)
    print(doc)
    print()
```

# Step 9 : Prompt Engineering

```python
context = "\n\n".join(results["documents"][0])

print(context)
```

```python
prompt = f"""
You are an experienced career assistant.

Your task is to generate a professional job application email.

Candidate Resume Information:
{context}

Job Description:
{job_description}

Instructions:

1. Write a professional job application email.
2. Highlight only the relevant skills and projects.
3. Do not invent any information.
4. Keep the email concise and professional.
5. Mention why the candidate is a good fit.

Return only the email.
"""
```

```python
response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents=prompt
)

print(response.text)
```

```python
import time

for attempt in range(3):
    try:
        response = client.models.generate_content(
            model="gemini-3.5-flash",
            contents=prompt
        )
        print(response.text)
        break
    except Exception as e:
        print(f"Attempt {attempt+1} failed:", e)
        time.sleep(10)
```
